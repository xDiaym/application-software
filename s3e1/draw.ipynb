```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
x = np.fromfile("out.dat", dtype=np.float32).reshape(-1, 102, 102)
```

```python

fig, ax = plt.subplots()
cax = ax.imshow(x[0], cmap='viridis', vmin=0, vmax=1)

def update(frame):
    cax.set_array(x[frame*100])
    return cax,

ani = FuncAnimation(fig, update, frames=x.shape[0]//100, interval=100)
ani.save('out.gif')
ani.save('out.mp4')
```
